```python
class Layer:

    def __init__(self, input_size, output_size, activation):
        """
        Represents a single neural network layer with forward and backward propagation capabilities.
        Parameters:
        - input_size (int): Size of the input vector.
        - output_size (int): Size of the output vector.
        - activation (function or str): Activation function to be used in this layer.
        """
        #initalize random weights
        self.weights = np.random.randn(input_size, output_size) * 0.01
        self.bias = np.zeros((1, output_size))
        self.activation = activation
        #define first and second moment
        self.w_m = np.zeros((input_size, output_size))
        self.w_v = np.zeros((input_size, output_size))
        self.b_m = np.zeros((1, output_size))
        self.b_v = np.zeros((1, output_size))

    def forward(self, input, train):
        """
        Performs forward propagation for this layer.
        Parameters:
        - input (np.array): Input data or activations from the previous layer.
        - train (bool): Flag indicating whether the model is in training mode.
        Returns:
        - np.array: Output of this layer after applying activation function.
        """
        self.input = input

        # Compute the linear transformation: z = W * input + b
        self.z = np.dot(input, self.weights) + self.bias

        # Apply the activation function to the linear transformation
        self.a = self.activation(self.z)
        return self.a

    def backward(self, da, lr, optimizer):
        """
        Performs backpropagation for this layer.
        Parameters:
        - da (np.array): Gradient of the loss with respect to the output of this layer.
        - lr (float): Learning rate.
        - optimizer (function): Optimizer function to update weights and biases.
        Returns:
        - np.array: Gradient of the loss with respect to the input of this layer.
        """
        # Special case handling for softmax activation
        if self.activation == "softmax":
            dz = da
        else:
            # Compute gradient of activation function
            dz = da * self.activation(self.z, derivative=True)
        # Compute gradients for weights and biases
        dw = np.dot(self.input.T, dz)
        db = np.sum(dz, axis=0)
        size = dz.shape[0]

        # Update weights and biases using the optimizer
        self.weights, self.w_m, self.w_v = optimizer(self.weights, (1/size) * dw, self.w_m, self.w_v)
        self.bias, self.b_m, self.b_v = optimizer(self.bias, (1/size) * db, self.b_m, self.b_v)
        return np.dot(dz, self.weights.T)
```

```python
class Adam():

    def __init__(self, learn_rate, b1, b2, epsilon=1e-8, time_step=0):
        """
        Initializes the Adam optimizer with hyperparameters.
        Parameters:
        - learn_rate (float): The learning rate.
        - b1 (float): Exponential decay rate for the first moment estimates.
        - b2 (float): Exponential decay rate for the second moment estimates.
        - epsilon (float): Small value to prevent division by zero.
        - time_step (int): Initial time step.
        """
        self.learn_rate = learn_rate
        self.b1 = b1
        self.b2 = b2
        self.epsilon = epsilon
        self.time_step = time_step  # Time step

    def __call__(self, weight, dloss, m, v):
        """
        Updates the weight using the Adam optimizer.
        Parameters:
        - weight (np.array): The current weight to be updated.
        - dloss (np.array): Gradient of the loss with respect to the weight.
        - m (np.array): First moment estimate.
        - v (np.array): Second moment estimate.
        Returns:
        - np.array: Updated weight.
        - np.array: Updated first moment estimate.
        - np.array: Updated second moment estimate.
        """
        self.time_step += 1
        m = self.b1 * m + (1 - self.b1) * dloss
        v = self.b1 * v + (1 - self.b1) * (dloss ** 2)

        # Bias-corrected first and second moment estimates
        m_hat = m / (1 - self.b1 ** self.time_step)
        v_hat = v / (1 - self.b2 ** self.time_step)

        # Adam update
        change = self.learn_rate * m_hat / (np.sqrt(v_hat) + self.epsilon)
        weight -= change

        return weight, m, v
```

```python
def relu(z, derivative=False):
    """
    Computes the Rectified Linear Unit (ReLU) activation function or its derivative.
    Parameters:
    - z (np.array): Input value or array.
    - derivative (bool): Flag indicating whether to compute the derivative of ReLU.
    Returns:
    - np.array: ReLU output or its derivative.
    """
    if derivative:
        return np.where(z > 0, 1, 0)
    return np.maximum(0, z)
```

```python
def softmax(z, derivative=True):
    """
    Computes the softmax activation function or its derivative.
    Parameters:
    - z (np.array): Input array, typically representing logits.
    - derivative (bool): Flag indicating whether to compute the derivative of softmax.
    Returns:
    - np.array: Softmax output or its derivative.
    """
    if derivative:
        # Compute the derivative of softmax
        # Subtract the maximum value for numerical stability
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
        softmax_output = exp_z / np.sum(exp_z, axis=1, keepdims=True)
        return np.clip(softmax_output, 1e-7, 1 - 1e-7)
```

```python
def cross_entropy_loss(y_pred, y_true, epsilon=1e-7):
    """
    Computes the cross-entropy loss and its gradient.
    Parameters:
    - y_pred (np.array): Predicted probabilities or logits.
    - y_true (np.array): True labels in one-hot encoded format.
    - epsilon (float): Small value added to predicted probabilities to avoid numerical instability.
    Returns:
    - float: Cross-entropy loss.
    - np.array: Gradient of the loss with respect to the predicted values.
    """
    m = y_true.shape[0]
    y_pred = np.clip(y_pred, epsilon, 1. - epsilon)
    loss = -np.sum(y_true * np.log(y_pred)) / m
    dy = y_pred - y_true
    return loss, dy
```

```python
class Dropout:

    def __init__(self, p):
        """
        Initializes the Dropout layer with a dropout probability.
        Parameters:
        - p (float): Dropout probability, representing the probability of dropping a unit.
        """
        self.p = p

    def forward(self, input, train=True):
        """
        Performs dropout during the forward pass.
        Parameters:
        - input (np.array): Input data or activations.
        - train (bool): Flag indicating whether to apply dropout during training.
        Returns:
        - np.array: Output after applying dropout.
        """
        if train:
            self.mask = np.random.rand(*input.shape) > self.p
            return input * self.mask / (1 - self.p)
        return input

    def backward(self, da, lr, optimizer):  # Add lr as an argument
        """
        Performs dropout during the backward pass.
        Parameters:
        - da (np.array): Gradient of the loss with respect to the output of this layer.
        Returns:
        - np.array: Gradient of the loss with respect to the input of this layer.
        """
        return da * self.mask / (1 - self.p)
```

```python
class BatchNorm:
    def __init__(self, input_size, alpha=0.9,epsilon=1e-5):
        """
        Initializes the Batch Normalization layer.
        Parameters:
        - input_size (int): Size of the input vector.
        - alpha (float): Exponential moving average parameter.
        - epsilon (float): Small value added to avoid division by zero.
        """
        self.gamma = np.ones((1, input_size))
        self.beta = np.zeros((1, input_size))
        self.epsilon = epsilon
        self.alpha = alpha
        self.running_mean = np.zeros((1, input_size))
        self.running_var = np.zeros((1, input_size))

        #initialize first and second moments
        self.g_m = np.ones((1, input_size))
        self.g_v = np.ones((1, input_size))
        self.b_m = np.zeros((1, input_size))
        self.b_v = np.zeros((1, input_size))

    def forward(self, input, train=True):
        """
        Performs batch normalization during the forward pass.
        Parameters:
        - input (np.array): Input data or activations.
        - train (bool): Flag indicating whether to apply batch normalization during training.
        Returns:
        - np.array: Output after applying batch normalization.
        """
        if train:
            # Compute batch mean and variance
            batch_mean = np.mean(input, axis=0)
            batch_var = np.var(input, axis=0)
            # Update running mean and variance using exponential moving averages
            self.running_mean = self.alpha * self.running_mean + (1 - self.alpha) * batch_mean
            self.running_var = self.alpha * self.running_var + (1 - self.alpha) * batch_var

            self.batch_mean = batch_mean
            self.batch_var = batch_var
            self.input_centered = input - batch_mean
            self.input_normalized = self.input_centered / np.sqrt(batch_var + self.epsilon)
            return self.gamma * self.input_normalized + self.beta
        else:
            # Normalize input using running mean and variance
            return self.gamma * (input - self.running_mean) / np.sqrt(self.running_var + self.epsilon) + self.beta

    def backward(self, da, lr, optimizer):
        """
        Performs backpropagation for batch normalization.
        Parameters:
        - da (np.array): Gradient of the loss with respect to the output of this layer.
        - lr (float): Learning rate.
        - optimizer (function): Optimizer function to update parameters.
        Returns:
        - np.array: Gradient of the loss with respect to the input of this layer.
        """
        m = da.shape[0]
        dx_normalized = da * self.gamma
        batch_var_clipped = np.clip(self.batch_var, 1e-7, np.inf)  # Clip batch_var
        dvar = -0.5 * np.sum(dx_normalized * self.input_centered, axis=0) * (batch_var_clipped + self.epsilon)**(-1.5)
        dmean = -np.sum(dx_normalized / np.sqrt(batch_var_clipped + self.epsilon), axis=0) - 2 * dvar * np.mean(self.input_centered, axis=0)
        dx = dx_normalized / np.sqrt(batch_var_clipped + self.epsilon) + dvar * 2 * self.input_centered / m + dmean / m
        dgamma = np.sum(da * self.input_normalized, axis=0)
        dbeta = np.sum(da, axis=0)

        # Update gamma and beta parameters using the optimizer
        self.gamma, self.g_m, self.g_v = optimizer(self.gamma, (1/m) * dgamma, self.g_m, self.g_v)
        self.beta, self.b_m, self.b_v = optimizer(self.beta, (1/m) * dbeta, self.b_m, self.b_v)
        return dx
```

```python
class NeuralNetwork:
    def __init__(self, layers):
        """
        Initializes the Neural Network with a list of layers.
        Parameters:
        - layers (list): List of Layer objects representing the network architecture.
        """
        self.layers = layers
        self.train_accuracies = []
        self.test_accuracies = []
        self.training_time=[]

    def forward(self, X, train=True):
        """
        Performs forward propagation through the network.
        Parameters:
        - X (np.array): Input data.
        - train (bool): Flag indicating whether to apply training-specific operations (e.g., dropout).
        Returns:
        - np.array: Output of the network.
        """
        for layer in self.layers:
            X = layer.forward(X, train)
        return X

    def backward(self, da, lr, optimizer):
        """
        Performs backward propagation through the network.
        Parameters:
        - da (np.array): Gradient of the loss with respect to the output of the network.
        - lr (float): Learning rate.
        - optimizer (function): Optimizer function to update parameters.
        """
        for layer in reversed(self.layers):
            da = layer.backward(da, lr, optimizer)

    def fit(self, X, y, epochs, batch_size, lr, optimizer, log=False, X_test=None, y_test=None):
        """
        Trains the neural network using mini-batch stochastic gradient descent.
        Parameters:
        - X (np.array): Training input data.
        - y (np.array): Training labels.
        - epochs (int): Number of training epochs.
        - batch_size (int): Size of mini-batches for training.
        - lr (float): Learning rate.
        - optimizer (function): Optimizer function to update parameters.
        - log (bool): Flag indicating whether to log training and test accuracies.
        - X_test (np.array): Test input data (optional).
        - y_test (np.array): Test labels (optional).
        """
        for epoch in range(epochs):
            indices = np.arange(X.shape[0])
            np.random.shuffle(indices)
            epo_start = time.time()
            for start in range(0, X.shape[0], batch_size):
                end = min(start + batch_size, X.shape[0])
                batch_indices = indices[start:end]
                X_batch, y_batch = X[batch_indices], y[batch_indices]
                y_pred = self.forward(X_batch)
                loss, dy = cross_entropy_loss(y_pred, y_batch)
                self.backward(dy, lr, optimizer)
            epo_time = time.time() - epo_start
            self.training_time.append(epo_time)
            if log:
                train_pred = self.predict(X)
                train_accuracy = np.mean(train_pred == np.argmax(y, axis=1))
                self.train_accuracies.append(train_accuracy)

                test_pred = self.predict(X_test)
                test_accuracy = np.mean(test_pred == np.argmax(y_test, axis=1))
                self.test_accuracies.append(test_accuracy)
                if (epoch+1)%5 == 0:
                    print(f"Epoch {epoch+1}/{epochs}, epoch time: {epo_time:.2f}s, Train accuracy: {train_accuracy:.4f}, Test accuracy: {test_accuracy:.4f}")


    def predict(self, X):
        """
        Predicts class labels for the input data.
        Parameters:
        - X (np.array): Input data for which predictions need to be made.
        Returns:
        - np.array: Predicted class labels.
        """
        return np.argmax(self.forward(X, train=False), axis=1)
```

```python
def make_layers(hidden_layers, dropout_rates, input_size, output_size):
    """
    Creates a list of neural network layers based on the specified architecture parameters.
    Parameters:
    - hidden_layers (list): List containing the sizes of hidden layers.
    - dropout_rates (list): List containing the dropout rates for each hidden layer.
    - input_size (int): Size of the input data.
    - output_size (int): Size of the output data.
    Returns:
    - list: List of neural network layers.
    """
    layers = []
    prev = input_size

    for i in range(len(hidden_layers)):
        size = hidden_layers[i]

        # Add a standard layer with ReLU activation
        layers.append(Layer(prev, size, relu))
        # Add batch normalization
        layers.append(BatchNorm(size))
        # Add dropout layer
        layers.append(Dropout(dropout_rates[i]))

        prev = size

    # Add the output layer with softmax activation
    layers.append(Layer(prev, output_size, softmax))

    return layers
```

```python
def evaluate_performance(model, X_test, y_test):
    """
    Evaluates the performance of a classification model using various evaluation metrics.
    Parameters:
    - model: The trained classification model.
    - X_test: Test input data.
    - y_test: True labels corresponding to the test input data.
    Returns:
    - dict: Dictionary containing evaluation metrics.
    """
    predictions = model.predict(X_test)
    #Turn into one hot
    y_test = np.argmax(y_test, axis=1)
    # True Positives, False Positives, True Negatives, False Negatives
    unique_classes = np.unique(y_test)
    precision_list = []
    recall_list = []
    f1_score_list = []
    specificity_list = []

    for label in unique_classes:
        TP = np.sum((predictions == y_test) & (predictions == label))
        FP = np.sum((predictions != y_test) & (predictions == label))
        FN = np.sum((predictions != y_test) & (y_test == label))
        TN = np.sum((predictions == y_test) & (predictions != label) & (y_test != label))

        precision = TP / (TP + FP) if TP + FP > 0 else 0
        recall = TP / (TP + FN) if TP + FN > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
        specificity = TN / (TN + FP) if TN + FP > 0 else 0

        precision_list.append(precision)
        recall_list.append(recall)
        f1_score_list.append(f1_score)
        specificity_list.append(specificity)

    # Macro-averaged metrics
    macro_precision = np.mean(precision_list)
    macro_recall = np.mean(recall_list)
    macro_f1_score = np.mean(f1_score_list)
    macro_specificity = np.mean(specificity_list)

    # Overall accuracy
    accuracy = np.mean(predictions == y_test)
    average_training_time = sum(model.training_time)/len(model.training_time)
    return {
        'precision': macro_precision,
        'recall': macro_recall,
        'f1_score': macro_f1_score,
        'specificity': macro_specificity,
        'accuracy': accuracy,
        'average_training_time':average_training_time
    }
```

Optimizer experiment

```python
class NoOptimizer():
    def __init__(self, learn_rate):
        self.learn_rate = learn_rate
    def __call__(self, weight, dloss, m, v):
        weight-=self.learn_rate*dloss
        return weight, m, v
```

```python
class Momentum():
    def __init__(self, learn_rate, gamma=0.9, decay=0):
        self.learn_rate = learn_rate
        self.gamma = gamma
        self.decay = decay

    def __call__(self, weight, dloss, m, v):
        '''calculate the weight update given the derivative of loss w.r.t. weight and previous momentum_term'''
        change = self.gamma * m + self.learn_rate * dloss
        weight = weight * (1 - self.decay) - change
        return weight, change, v
```

```python
# effect of adam
# Define the model architecture
lr = 0.0001
epochs=30
batch_size = 64
hidden_layers = [256, 128]
dropout_rates = [0.2,0.3]
optimizer_0 = Adam(0.0001, 0.9, 0.999)#base parameters
optimizer_1 = NoOptimizer(0.0001)#no optimizer
optimizer_2 = Adam(0.0001, 0.5, 0.999)#lower beta1
optimizer_3 = Adam(0.0001, 0.99, 0.999)#higher beta1
optimizer_4 = Adam(0.0001, 0.9, 0.8)#lower beta2
optimizer_5 = Adam(0.001, 0.9, 0.999)#high learning rate
optimizer_6 = Momentum(0.0001, 0.9, 0.0001)#Momentum

model_0 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_1 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_2 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_3 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_4 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_5 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))
model_6 = NeuralNetwork(make_layers(hidden_layers, dropout_rates, input_size, num_classes))

models = [(model_0, 'base parameters', optimizer_0), (model_1, 'No adam', optimizer_1), (model_2, 'lower beta1', optimizer_2),
           (model_3, 'higher beta1', optimizer_3), (model_4, 'lower beta2', optimizer_4),
           (model_5, 'high learning rate', optimizer_5), (model_6, 'momentum', optimizer_6)]
```
